```python
from quadtree2d import *
```

```python
# Generate a root with depth 0 and value -1
root = Node2D([100,100,200,200] ,0 ,-1)
```

```python
# split the root with four children
root.split()
# get the number of leaves
print('the number of leaves', len(root.get_leaves()))
# remove the children and generate a pecfect subquadtree from it
root.children = []
root.generate_perfect_subquadtree(4)
print('the number of leaves', len(root.get_leaves()))
```

```python
# remove children and use the root to build a quadtree
root.children = []
quadtree = QuadTree2D(root, 5)
quadtree.plot()
```

```python
quadtree.explore_update([127,139],0)
quadtree.plot()
```

```python
quadtree.explore_update([177,169],1)
quadtree.plot()
```
